```python
import os
import calendar
import pandas as pd
import time

cal = calendar.Calendar()

qlist = ["여성", "여자+-여성", "남성", '남자++남성']

count = 0

for query in qlist:
    query1 = "#3_" + query
    for year in range(2000, 2020):
        for month in range(1, 13):
                monthdays = [d for d in cal.itermonthdays(year, month) if d != 0]
                for day in monthdays:

                    date = str(year) + str(month).zfill(2) + str(day).zfill(2)
                    print(date)
                    
                    xlsx_name = os.path.join(os.path.dirname(os.getcwd()), 'xlsx_data', 
                                              '#3_knu_score', query1, '#3_'+ str(year) + '_'+query, 
                                             '{}.xlsx'.format("#3_"+date+"_"+query))

                    df = pd.read_excel(xlsx_name)
                    count += len(df)

top = count // 10
print(count, top)
```

```python
start = time.time()
print("hello")

pos_score, neg_score = list(), list()
df_pos = pd.DataFrame(columns=['title', 'source', 'date', 'text', 'knu_score'])
df_neg = pd.DataFrame(columns=['title', 'source', 'date', 'text', 'knu_score'])
df_pos_name = os.path.join(os.path.dirname(os.getcwd()), 'xlsx_data', '#3_knu_score', f'{positive}.xlsx')
df_neg_name = os.path.join(os.path.dirname(os.getcwd()), 'xlsx_data', '#3_knu_score', f'{negative}.xlsx')

for query in qlist:
    query1 = "#3_" + query
    for year in range(2000, 2020):
        for month in range(1, 13):
                monthdays = [d for d in cal.itermonthdays(year, month) if d != 0]
                for day in monthdays:

                    date = str(year) + str(month).zfill(2) + str(day).zfill(2)
                    print(date)
                    
                    xlsx_name = os.path.join(os.path.dirname(os.getcwd()), 'xlsx_data', 
                                              '#3_knu_score', query1, '#3_'+ str(year) + '_'+query, 
                                             '{}.xlsx'.format("#3_"+date+"_"+query))

                    df = pd.read_excel(xlsx_name)

                    for i in range(len(df)):
                        
                        d = df.loc[i]
                        k = d['knu_score']
                        
                        if k > 0 and k > min(pos_score):
                            pos_score.append(k)
                            df_pos.append(d)
                        if len(pos_score) > top:
                            ind = pos_score.index(min(pos_score))
                            pos_score.pop(ind)
                            df_pos.drop(ind)
                            df_pos.reset_index(drop=True)
                            
                        if k < 0 and k < max(neg_score):
                            neg_score.append(k)
                            df_neg.append(d)
                        if len(neg_score) > top:
                            ind = neg_score.index(max(neg_score))
                            neg_score.pop(ind)
                            df_neg.drop(ind)
                            df_neg.reset_index(drop=True)
                            
df_pos.to_excel(df_pos_name)
df_neg.to_excel(df_neg_name)

end = time.time()
print(end - start)
```
