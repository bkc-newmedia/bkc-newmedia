```python
import os
import glob
import pandas as pd

```

```python
# 년도별
#남
mans=['남성','남자+-남성']

cat0_sent1_m, cat1_sent1_m, cat2_sent1_m, cat3_sent1_m, cat4_sent1_m, cat5_sent1_m, cat6_sent1_m=0,0,0,0,0,0,0
cat0_sent0_m, cat1_sent0_m, cat2_sent0_m, cat3_sent0_m, cat4_sent0_m, cat5_sent0_m, cat6_sent0_m=0,0,0,0,0,0,0


    

for year in range(2000,2020):
    df_m = pd.DataFrame(columns=['title','source','date','url','text','sentiment'])

    for month in range(1,13):
        for man in mans:
            path='C:/Users/joo/Desktop/#4_남성/#4_'+str(year)+'_'+str(month).zfill(2)+'_'+man+'.xlsx'
            print(path)
            df_t=pd.read_excel(path)
            df_m=df_m.append(df_t)
            
    cat0_sent0_m+=len(df_m[(df_m['category']==0 )& (df_m['sentiment']==0)])
    cat0_sent1_m+=len(df_m[(df_m['category']==0 )& (df_m['sentiment']==1)])
    cat1_sent0_m+=len(df_m[(df_m['category']==1 )& (df_m['sentiment']==0)])
    cat1_sent1_m+=len(df_m[(df_m['category']==1 )& (df_m['sentiment']==1)])
    cat2_sent0_m+=len(df_m[(df_m['category']==2 )& (df_m['sentiment']==0)])
    cat2_sent1_m+=len(df_m[(df_m['category']==2 )& (df_m['sentiment']==1)])
    cat3_sent0_m+=len(df_m[(df_m['category']==3 )& (df_m['sentiment']==0)])
    cat3_sent1_m+=len(df_m[(df_m['category']==3 )& (df_m['sentiment']==1)])
    cat4_sent0_m+=len(df_m[(df_m['category']==4 )& (df_m['sentiment']==0)])
    cat4_sent1_m+=len(df_m[(df_m['category']==4 )& (df_m['sentiment']==1)])
    cat5_sent0_m+=len(df_m[(df_m['category']==5 )& (df_m['sentiment']==0)])
    cat5_sent1_m+=len(df_m[(df_m['category']==5 )& (df_m['sentiment']==1)])
    cat6_sent0_m+=len(df_m[(df_m['category']==6 )& (df_m['sentiment']==0)])
    cat6_sent1_m+=len(df_m[(df_m['category']==6 )& (df_m['sentiment']==1)])


        
```

```python
pos_m=[cat0_sent1_m, cat1_sent1_m, cat2_sent1_m, cat3_sent1_m, cat4_sent1_m, cat5_sent1_m, cat6_sent1_m]
neg_m=[cat0_sent0_m, cat1_sent0_m, cat2_sent0_m, cat3_sent0_m, cat4_sent0_m, cat5_sent0_m, cat6_sent0_m]
percent_pos_m=[]
percent_neg_m=[]
for i in range(7): # 7인데
    percent_pos_m.append((pos_m[i]/(pos_m[i]+neg_m[i]))*100)
    percent_neg_m.append((neg_m[i]/(pos_m[i]+neg_m[i]))*100)
```

```python
import pygal


py_bar = pygal.HorizontalStackedBar(style=pygal.style.styles['default'](label_font_size=10))

py_bar.x_labels = ['세계','정치','IT과학','연예','생활문화','경제','사회']

py_bar.add("긍정",pos_m)
py_bar.add("부정",neg_m)


py_bar.render_to_file("category_man.svg")
py_bar
```

```python
import pygal


py_bar = pygal.HorizontalStackedBar(style=pygal.style.styles['default'](label_font_size=10))
py_bar.x_labels = ['세계','정치','IT과학','연예','생활문화','경제','사회']

percent_formatter = lambda x: '{:.2f}%'.format(x)
py_bar.value_formatter = percent_formatter


py_bar.add("긍정",percent_pos_m)
py_bar.add("부정",percent_neg_m)


py_bar.render_to_file("category_man_percent.svg")
py_bar
```
